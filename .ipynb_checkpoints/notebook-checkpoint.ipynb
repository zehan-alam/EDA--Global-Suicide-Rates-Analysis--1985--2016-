# Importing Data and Libraries

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
import ipywidgets as widgets
```

```python
df = pd.read_csv('Datasets/Suicide Rates Overview 1985-2016.csv')
df = df[df['year']!=2016]
display(df.head())
df.info()
```

```python
display(df.describe())
display(df.describe(include=['O']))
```

# Data Cleanup

```python
(df.isnull().sum()/len(df))*100
```

```python
df = df.drop(columns=['country-year','HDI for year']).copy()
```

```python
df[' gdp_for_year ($) '] = df[' gdp_for_year ($) '].str.replace(',','').astype('int')
df[' gdp_for_year ($) '] = (df[' gdp_for_year ($) ']/1000000000).round(3)
```

```python
df['sex'] = df['sex'].str.capitalize()
```

```python
df = df.rename(columns={'country':'Country','year':'Year','sex':'Sex', 'age':'Age Group', 'suicides_no':'No. of Suicides', 
                        'population':'Population','suicides/100k pop':'Suicides/100k pop', ' gdp_for_year ($) ':'GDP(B$)',
                        'gdp_per_capita ($)':'GDP per Capita($)','generation':'Generation'}).copy()
```

```python
display(df.head())
df.info()
```

# Data Analysis w/ Report

```python
display(df.describe().loc[['mean','std','min','max']])
display(df.describe(include=['O']).loc[['unique']])
```

```python
sns.heatmap(df.select_dtypes(exclude='object').corr()*100,annot=True)
```

## 1. Yearly Stats

```python
pop_yearly = df.groupby('Year')['Population'].sum()
plt.figure(figsize=(12,5))
g1 = sns.lineplot(x=pop_yearly.index,y=pop_yearly.values)
g1.set_title('Recorded Population Each Year')
g1.set_ylabel('Population')
g1.set_yticklabels(['{:,.2f}'.format(y) + 'B' for y in g1.get_yticks()/1000000000])
plt.show()
```

```python
suicide_yearly = df.groupby('Year')['No. of Suicides'].sum()
plt.figure(figsize=(12,5))
g2 = sns.lineplot(x=suicide_yearly.index,y=suicide_yearly.values)
g2.set_title('Total No. of Suicide Each Year')
g2.set_yticklabels(['{:,.0f}'.format(y) + 'K' for y in g2.get_yticks()/1000])
g2.set_ylabel('No. of Suicides')
plt.show()
```

```python
suicide_yearly_per_100k = df.groupby('Year')['Suicides/100k pop'].mean()
plt.figure(figsize=(12,5))
g3 = sns.lineplot(x=suicide_yearly_per_100k.index,y=suicide_yearly_per_100k.values)
g3.set_title('Average Suicides/100k pop/100k Population Each Year')
g3.set_ylabel('Suicides/100k pop')
plt.show()
```

## Country

```python
top10_suicide_countries = df.groupby(['Country'])['No. of Suicides'].sum().sort_values(ascending=False)[:10].index
suicide_countries_sex = pd.DataFrame(df[df['Country'].isin(top10_suicide_countries)].groupby(['Country','Sex'])
                                    ['No. of Suicides'].sum()).reset_index()
suicide_countries_sex_cross = suicide_countries_sex.pivot_table(index='Country',columns='Sex',values='No. of Suicides')
suicide_countries_sex_cross['Total'] = suicide_countries_sex_cross['Male']+suicide_countries_sex_cross['Female']
suicide_countries_sex_cross = suicide_countries_sex_cross.sort_values('Total',ascending=False)

plt.figure(figsize=(12,5))
plt.bar(suicide_countries_sex_cross.index, suicide_countries_sex_cross['Male'],label='Male')
plt.bar(suicide_countries_sex_cross.index,suicide_countries_sex_cross['Female'], bottom=suicide_countries_sex_cross['Male'],label='Female',color='pink')
plt.title('Top 10 Countries with Most Suicides Recorded(1985-2015)')
plt.gca().set_yticklabels(['{:,.2f}'.format(y) + 'M' for y in plt.gca().get_yticks()/1000000])
plt.ylabel('No. of Suicides')
for i in range(10):
    plt.gca().text(x=i, y=suicide_countries_sex_cross['Female'].iloc[i]+suicide_countries_sex_cross['Male'].iloc[i]+10000, 
                   s=f'{round((suicide_countries_sex_cross["Female"].iloc[i]+suicide_countries_sex_cross["Male"].iloc[i])/1000000,2)}M',
                   ha='center', color='black')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
top5_country_suicide = pd.DataFrame(df[df['Country'].isin(top10_suicide_countries[:5])].groupby(['Year', 'Country'])['No. of Suicides'].sum()).reset_index()
plt.figure(figsize=(12,6))
g5 = sns.lineplot(data=top5_country_suicide, x='Year', y='No. of Suicides', hue='Country',palette='husl')
g5.set_title('Top 5 Countries with Highest No. of Suicides Over the Years')
g5.set_yticklabels(['{:,.0f}'.format(y) + 'K' for y in g5.get_yticks()/1000])
plt.show()
```

```python
top10_suicide_countries_100k = df.groupby(['Country'])['Suicides/100k pop'].mean().sort_values(ascending=False)[:10].index
suicide_countries_sex_100k = pd.DataFrame(df[df['Country'].isin(top10_suicide_countries_100k)].groupby(['Country','Sex'])
                                    ['Suicides/100k pop'].mean()).reset_index()
suicide_countries_sex_cross = suicide_countries_sex_100k.pivot_table(index='Country',columns='Sex',values='Suicides/100k pop')
suicide_countries_sex_cross['Total'] = (suicide_countries_sex_cross['Male']+suicide_countries_sex_cross['Female'])/2
suicide_countries_sex_cross = suicide_countries_sex_cross.sort_values('Total',ascending=False)

plt.figure(figsize=(12,5))
plt.bar(suicide_countries_sex_cross.index, suicide_countries_sex_cross['Male'],label='Male')
plt.bar(suicide_countries_sex_cross.index,suicide_countries_sex_cross['Female'], bottom=suicide_countries_sex_cross['Male'],label='Female',color='pink')
plt.title('Top 10 Countries with Most Suicides/100k pop(1985-2015)')
plt.ylabel('Suicides/100k pop')
for i in range(10):
    plt.gca().text(x=i, y=suicide_countries_sex_cross['Female'].iloc[i]+suicide_countries_sex_cross['Male'].iloc[i]+1, 
                   s=round((suicide_countries_sex_cross["Female"].iloc[i]+suicide_countries_sex_cross["Male"].iloc[i])/2,2),
                   ha='center', color='black')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
top5_country_suicide_100k = pd.DataFrame(df[df['Country'].isin(top10_suicide_countries_100k[:5])].groupby(['Year', 'Country'])['Suicides/100k pop'].mean()).reset_index()
plt.figure(figsize=(12,7))
sns.lineplot(data=top5_country_suicide_100k, x='Year', y='Suicides/100k pop', hue='Country',palette='husl')
plt.title('Top 5 Countries with Highest Suicide per 100k People Over the Years')
plt.show()
```

## Age

## GDP

```python
fig,ax = plt.subplots(1,2,figsize=(16,6))

gdp_100k = df.groupby(['Country','Year','Sex']).agg({'Suicides/100k pop':'mean','GDP per Capita($)':'mean'}).reset_index()
sns.scatterplot(data=gdp_100k,x='GDP per Capita($)',y='Suicides/100k pop',hue='Sex',palette=['pink','#2986cc'],ax=ax[0])
ax[0].set_title('Sucides/100K vs GDP per Capita')

gdp_total = df.groupby(['Country','Year','Sex']).agg({'No. of Suicides':'sum','GDP per Capita($)':'mean'}).reset_index()
sns.scatterplot(data=gdp_total,x='GDP per Capita($)',y='No. of Suicides',hue='Sex',palette=['pink','#2986cc'],ax=ax[1])
ax[1].set_title('Total No. of Suicide vs GDP per Capita')
plt.show()
```

```python
gdp_all = df.groupby('Year').agg(
    {'Population':'sum','Suicides/100k pop':'mean','GDP per Capita($)':'mean'})

fig,ax = plt.subplots(1,3,figsize=(16,5))
ax = ax.flatten()
for i,col in enumerate(gdp_all.columns):
    sns.lineplot(data=gdp_all, x='Year',y=col,ax=ax[i])
    ax[i].set_title(col)
ax[0].set_yticklabels(['{:,.2f}'.format(y) + 'B' for y in ax[0].get_yticks()/1000000000])
fig.tight_layout()
```

```python
country = 'Ukraine'
df_country = df[df['Country']==country].groupby('Year').agg(
    {'Population':'sum','Suicides/100k pop':'mean','GDP per Capita($)':'mean'}).reset_index()

fig = make_subplots(rows=1, cols=3, subplot_titles=("Population", "Suicides/100k Population", "GDP per Capita($)"))

trace1 = go.Scatter(x=df_country['Year'], y=df_country['Population'], line={'color':'#2986cc'}, name='')
trace2 = go.Scatter(x=df_country['Year'], y=df_country['Suicides/100k pop'], line={'color':'#2986cc'}, name='')
trace3 = go.Scatter(x=df_country['Year'], y=df_country['GDP per Capita($)'], line={'color':'#2986cc'}, name='')

fig.add_trace(trace1, row=1, col=1)
fig.add_trace(trace2, row=1, col=2)
fig.add_trace(trace3, row=1, col=3)
layout = go.Layout(
    xaxis={'title':"X Label"},
    yaxis={'title':"Y label"}) 
fig.update_layout(showlegend=False, template='simple_white', xaxis={'title':"Year"},xaxis2={'title':'Year'},xaxis3={'title':'Year'},
                  yaxis={'title':"Population"},yaxis2={'title':'Suicides/100k pop'},yaxis3={'title':'GDP per Capita($)'})

fig.show()
```

```python
fig = make_subplots(rows=1, cols=3, subplot_titles=("Population", "Suicides/100k Population", "GDP per Capita($)"))
def update_plots(country):
    df_country = df[df['Country'] == country].groupby('Year').agg(
        {'Population': 'sum', 'Suicides/100k pop': 'mean', 'GDP per Capita($)': 'mean'}).reset_index()

    trace1 = go.Scatter(x=df_country['Year'], y=df_country['Population'], line={'color': '#2986cc'}, name='')
    trace2 = go.Scatter(x=df_country['Year'], y=df_country['Suicides/100k pop'], line={'color': '#2986cc'}, name='')
    trace3 = go.Scatter(x=df_country['Year'], y=df_country['GDP per Capita($)'], line={'color': '#2986cc'}, name='')

    fig.add_trace(trace1, row=1, col=1)
    fig.add_trace(trace2, row=1, col=2)
    fig.add_trace(trace3, row=1, col=3)
    
    fig.update_layout(title_text=country,showlegend=False, template='simple_white', xaxis={'title':"Year"},
                      xaxis2={'title':'Year'},xaxis3={'title':'Year'},yaxis={'title':"Population"},yaxis2={'title':'Suicides/100k pop'},
                      yaxis3={'title':'GDP per Capita($)'})
    fig.show()

country_selector = widgets.Dropdown(
    options=df['Country'].unique(),
    value='Albania',
    description='Select Country:'
)
widgets.interactive(update_plots, country=country_selector)
```
